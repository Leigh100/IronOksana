```python
import numpy as np
import pandas as pd

df = pd.read_csv('Monday.csv')
df
```

1. Replacing null values – Replace missing values with means of the column (for numerical columns). Pay attention that the Income feature for instance has 0s which is equivalent to null values. (We assume here that there is no such income with 0 as it refers to missing values) Hint: numpy.nan is considered of float64 data type.

```python
customer_lifetime_value = df["customer lifetime value"].mean() #Calculating mean for the column
customer_lifetime_value
```

```python
income = df["income"].mean() #Calculating mean for the column
income
```

```python
number_of_open_complaints= df["number of open complaints"].mean() #Calculating mean for the column
number_of_open_complaints
```

```python
total_claim_amount = df['total claim amount'].mean() #Calculating mean for the column
total_claim_amount
```

```python
df.isna().sum() #Checking number of Nan
```

```python
df.isin([0]).sum(axis=0) #Checking number of '0'
```

```python
df['customer lifetime value'] = df['customer lifetime value'].replace(np.nan, customer_lifetime_value)
df['income'] = df['income'].replace(np.nan, income)
df['number of open complaints'] = df['number of open complaints'].replace(np.nan, number_of_open_complaints)
df['total claim amount'] = df['total claim amount'].replace(np.nan, total_claim_amount)

df
```

```python
df.isna().sum() #Double checking if there is no anymore NaN in numeric columns
```

```python
df['income'] = df['income'].replace(0, income) #Replacing '0' in income column with the Mean of the column
df
```

```python
df.isin([0]).sum(axis=0) #Checking if income column does not have anymore '0' values.
```

2. Bucketing the data - Write a function to replace column "State" to different zones. California as West Region, Oregon as North West, and Washington as East, and Arizona and Nevada as Central

```python
grouping_lists = [['California'], ['Oregon'], 
                  ['Washington'], ['Arizona', 'Nevada']] 
names =  ['West Region', 'North Region', 'East', 'Central']

# create a df with mapping information
maps = (pd.DataFrame({'zone': names, 'st': grouping_lists})
        .explode('st')
        .reset_index(drop=True))

# join maps
df = df.merge(maps, on = 'st', how='left').fillna("Other")

df
```

```python
df.to_csv('Tuesday.csv', index =False)
```
