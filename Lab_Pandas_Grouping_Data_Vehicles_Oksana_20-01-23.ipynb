Converting Grams/Mile to Grams/Km

1 Mile = 1.60934 Km

Converting Gallons to Liters

1 Gallon = 3.78541 Liters

What brand has the most cars?

```python
#Changing the column name
cars. rename(columns = {'CO2 Emission Grams/Mile':'CO2 Emission Grams/Km'}, inplace = True) 
cars.rename(columns = {'City MPG':'City_KMpL', 'Highway MPG':'Highway_KMpL', 'Combined MPG':'Combined_KMpL' }, inplace = True)
cars
```

```python
#Converting Grams/Mile to Grams/Km
cars['CO2 Emission Grams/Km'] = cars['CO2 Emission Grams/Km'].apply(lambda x: x / 1.60934) 
cars
```

```python
# Converting Gallons to Liters
# 1 Gallon = 3.78541 Liters
cars['City_KMpL'] = cars['City_KMpL'].apply(lambda x: x /2.352)
cars['Highway_KMpL'] = cars['Highway_KMpL'].apply(lambda x: x /2.352)
cars['Combined_KMpL'] = cars['Combined_KMpL'].apply(lambda x: x /2.352)
cars
```

```python
# What brand has the most cars?

cars['Make'].describe()  #cars.groupby('Make')['Model'].count().nlargest()  - another way
```

```python
cars['Make'].value_counts()
```

<b>show the average CO2_Emission_Grams/Km  by Brand

```python
cars.groupby("Make")[["CO2 Emission Grams/Km"]].mean()  
#cars.groupby("Make")[["CO2 Emission Grams/Km"]].apply(lambda x: x.mean()/1.60934) 
# another way 
```

<b>show the average CO2_Emission_Grams/Km  by Brand ... sorted

```python
cars.groupby("Make")[["CO2 Emission Grams/Km"]].mean().sort_values(by="CO2 Emission Grams/Km")
```

# (Optional) 

Use `pd.cut` or `pd.qcut` to create 4 groups (bins) of cars, by Year. We want to explore how cars have evolved decade by decade.

```python
cars['Year'].describe()
```

```python
bin_labels = ['80-s', '90-s','00-s', '10-s'] 

cars['Grouped_Year'] = pd.qcut(cars['Year'],
                              q=4,
                              labels=bin_labels)
cars
```

### Did cars consume more gas in the eighties?

show the average City_Km/Liter by year_range

```python
cars.groupby("Grouped_Year")["City_KMpL"].mean()
```

Which brands are more environment friendly?

```python
cars.groupby(["Grouped_Year", 'Make'])["CO2 Emission Grams/Km"].mean()
```

Does the drivetrain affect fuel consumption?

```python

cars.groupby("Drivetrain").agg({"City_KMpL":["mean"],"Highway_KMpL":"mean"})
```
