```python
import requests # to download html code
from bs4 import BeautifulSoup # to navigate through the html code
import pandas as pd
import numpy as np
import webbrowser
import random
from IPython.display import IFrame
import config
import spotipy
import json
from spotipy.oauth2 import SpotifyClientCredentials
from sklearn.compose import ColumnTransformer
from sklearn.preprocessing import StandardScaler
from sklearn.cluster import KMeans



#Initialize SpotiPy with user credentias
sp = spotipy.Spotify(auth_manager=SpotifyClientCredentials(client_id= config.client_id,
                                                           client_secret= config.client_secret))
```

```python
df = pd.read_csv("hot_music.csv")
df
```

```python
songs= list(df.song.values)
songs
```

```python
lowercase_list = [s.lower() for s in songs]
print(lowercase_list)
```

```python
df1 = pd.read_csv('9944_songs.csv', index_col=0)
df1
```

```python
features = ['danceability', 'energy', 'loudness', 'mode', 'speechiness', 'acousticness', 'instrumentalness', 'liveness', 'valence', 'tempo']
kmeans = KMeans(n_clusters=7, random_state=0).fit(df1[features])
df1['cluster'] = kmeans.labels_
df1
```

```python
# Asking user for his fav song
favorite_song = input("What is your favorite song? ").strip()

# Check if the song is in the lowercase_list
if favorite_song.lower() in lowercase_list:
    # If the song is in the list, find its Spotify track ID
    search_url = 'https://open.spotify.com/search/' + favorite_song.replace(' ', '%20')
    response = requests.get(search_url)
    soup = BeautifulSoup(response.content, 'html.parser')
    results = sp.search(q=favorite_song,limit=3,market="GB")
    track_id=results["tracks"]["items"][0]["id"]
        
    # Play the song using IFrame
    iframe_url = f'https://open.spotify.com/embed/track/{track_id}'
    display(IFrame(src=iframe_url, width='300', height='380'))
else:
    results = sp.search(q=favorite_song, limit=3, market="GB")
    track_id=results["tracks"]["items"][0]["id"]
    features_user_fav_song= sp.audio_features(results["tracks"]["items"][0]["id"])
    df_user_fav_song = pd.DataFrame(features_user_fav_song)
    user_fav_song_cluster = kmeans.predict(df_user_fav_song[features])[0]
    recommendations = df1[df1['cluster'] == user_fav_song_cluster].sample(n=1)
    recommended_song_id = recommendations['id'].values[0]
    print("We recommend you the following song:")
    
    iframe_url = f'https://open.spotify.com/embed/track/{recommended_song_id}'
    display(IFrame(src=iframe_url, width='300', height='380'))
```
